```python
with open("input.txt", "r") as file:
    data = file.read()
```

```python
def check_doublet(s) -> bool:
    mid = len(s) // 2
    return s[:mid] == s[mid:]

total_sum = 0
for ran in data.split(","):
    start, end = map(int, ran.split("-"))
    if len(str(start)) == len(str(end)) and len(str(start)) % 2 == 1:
        continue
    for num in range(start, end + 1):
        if len(str(num)) % 2 == 0 and check_doublet(str(num)):
            total_sum += num
print(total_sum)
```

```python
def check_reps(s) -> bool:
    if len(s) < 2:
        return False
    # string consists of repeats only, like 11, 111, 121212, etc.
    mid = len(s) // 2
    for i in range(1, mid+1):
        if len(s) % i != 0:
            continue
        # split str into chunks of size i and compare
        all_match = True
        for j in range(0, len(s), i):
            if s[j:j+i] != s[0:i]:
                all_match = False
                break
        if all_match:
            return True
    return False

total_sum = 0
for ran in data.split(","):
    start, end = map(int, ran.split("-"))
    for num in range(start, end + 1):
        if check_reps(str(num)):
            total_sum += num
print(total_sum)
```

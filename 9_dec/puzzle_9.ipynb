```python
with open("input.txt") as f:
    coords = np.array([list(map(int, line.strip().split(",")))
                        for line in f if line.strip()], dtype=np.int64)
n = len(coords)

# build polygon edges between consecutive corners (wrap-around)
# store each edge as (x_lo, y_lo, x_hi, y_hi) i.e. bounding box of the (axis-aligned) segment
prev = np.roll(coords, 1, axis=0)
lo = np.minimum(coords, prev)
hi = np.maximum(coords, prev)
ex_lo, ey_lo = lo[:, 0], lo[:, 1]
ex_hi, ey_hi = hi[:, 0], hi[:, 1]

max_s = 0

for i in range(n):
    x1, y1 = coords[i]
    for j in range(i + 1, n):
        x2, y2 = coords[j]

        # opposite corners must differ in both x and y
        if x1 == x2 or y1 == y2:
            continue

        xL, xR = (x1, x2) if x1 < x2 else (x2, x1)
        yB, yT = (y1, y2) if y1 < y2 else (y2, y1)

        # reject if ANY polygon edge passes through the STRICT interior of the rectangle
        # (touching the rectangle border is allowed -> strict inequalities)
        crosses_interior = np.any(
            (ex_hi > xL) & (ex_lo < xR) &
            (ey_hi > yB) & (ey_lo < yT)
        )
        if crosses_interior:
            continue

        # valid rectangle (all tiles are red/green)
        s = (xR - xL + 1) * (yT - yB + 1)
        if s > max_s:
            max_s = s

print(int(max_s))
```

```python
with open("input.txt", "r") as file:
    data = file.read()
```

```python
total_sum = 0
for line in data.splitlines():
    top_1st = 0
    top_2nd = 0
    for i, c in enumerate(line):
        if int(c) > top_1st and i < len(line.strip()) -1:
            top_2nd = 0
            top_1st = int(c)
        elif int(c) > top_2nd and i != 0:
            top_2nd = int(c)

    total_sum += top_1st*10 + top_2nd
total_sum
```

```python
to_pick = 12
total_sum_12 = 0
for line in data.splitlines():
    digits = [int(c) for c in line.strip()]
    n = len(digits)
    dp = [[-1] * (to_pick + 1) for _ in range(n + 1)]
    
    for i in range(n + 1):
        dp[i][0] = 0
    
    for i in range(1, n + 1):
        for j in range(1, min(i, to_pick) + 1):
            dp[i][j] = max(dp[i - 1][j], dp[i - 1][j - 1] * 10 + digits[i - 1])
    
    total_sum_12 += dp[n][to_pick]
total_sum_12
```

```python
import numpy as np
```

```python
with open("input.txt") as f:
    input = [line.rstrip("\n").split(',') for line in f]
```

```python
coords = []
for i in input:
    coords.append([int(x) for x in i])
coords = np.array(coords)
```

```python
# i will calculate distances i want to keep 1000 shortest distances
distances = []
for i in range(len(coords)):
    for j in range(i+1, len(coords)):
        distances.append((np.sqrt(np.sum((coords[i] - coords[j])**2)), i, j))
distances.sort(key=lambda x: x[0])
distances = np.array(distances[:1000])[:, 1:].astype(int)
```

```python
def cluster_pairs(pairs):
    clusters = []

    for a, b in pairs:
        hits = []

        # find clusters that contain a or b
        for c in clusters:
            if a in c or b in c:
                hits.append(c)

        if not hits:
            # new cluster
            clusters.append({a, b})
        else:
            # merge all hit clusters + (a, b)
            merged = {a, b}
            for c in hits:
                merged |= c
                clusters.remove(c)
            clusters.append(merged)

    return clusters

clusters = cluster_pairs(distances)
```

```python
cluster_sizes = [len(c) for c in clusters]
top3 = sorted(cluster_sizes, reverse=True)[:3]
answer = 1
for size in top3:
    answer *= size
answer
```

```python
distances = []
for i in range(len(coords)):
    for j in range(i + 1, len(coords)):
        d = np.sqrt(np.sum((coords[i] - coords[j]) ** 2))
        distances.append((d, i, j))

distances.sort(key=lambda x: x[0])
```

```python
def find_final_connection(distances, coords):
    clusters = []
    num_nodes = len(coords)

    for _, a, b in distances:
        hits = []

        for c in clusters:
            if a in c or b in c:
                hits.append(c)

        if not hits:
            clusters.append({a, b})
        else:
            merged = {a, b}
            for c in hits:
                merged |= c
                clusters.remove(c)
            clusters.append(merged)

        if len(clusters) == 1 and len(clusters[0]) == num_nodes:
            return coords[a][0] * coords[b][0]
print(find_final_connection(distances, coords))
```

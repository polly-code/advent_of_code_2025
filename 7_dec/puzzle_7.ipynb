```python
with open("input.txt") as f:
    input = [line.rstrip("\n") for line in f]
```

```python
answer = 0
cur_beams = set(i for i, c in enumerate(input[0]) if c == 'S')
for line in input[1:]:
    new_beams = set()
    for beam in cur_beams:
        if line[beam] == '^':
            answer += 1
            new_beams.add(beam-1)
            new_beams.add(beam+1)
        else:
            new_beams.add(beam)
    cur_beams = new_beams
answer
```

```python
cur_beams = [1 if c == 'S' else 0 for c in input[0]]
for line in input[1:]:
    new_beams = [0] * len(cur_beams)
    for i, beam in enumerate(cur_beams):
        if beam:
            if line[i] == '^':
                new_beams[i-1] += beam
                new_beams[i+1] += beam
            else:
                new_beams[i] += beam
    cur_beams = new_beams
sum(cur_beams)
```

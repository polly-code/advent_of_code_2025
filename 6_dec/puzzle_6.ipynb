```python
with open("input.txt") as f:
    input = f.readlines()
```

```python
answer = 0
m = []
for line in input[:-1]:
    line = line.strip().split()
    for i,c in enumerate(line):
        try:
            m[i].append(int(c))
        except IndexError:
            m.append([int(c)])
for i, c in enumerate(input[-1].strip().split()):
    if c == '+':
        answer += sum(m[i])
    elif c == '*':
        prod = 1
        for n in m[i]:
            prod *= n
        answer += prod
print(answer)
```

```python
height = len(input) - 1
width = max(len(line) for line in input)

grid = [line.ljust(width) for line in input]
blank_col = [
    all(grid[row][col] == " " for row in range(len(grid)))
    for col in range(width)
]

answer = 0
col = 0

while col < width:
    if blank_col[col]:
        col += 1
        continue

    start = col
    while col < width and not blank_col[col]:
        col += 1
    end = col - 1

    ops = {grid[height][c] for c in range(start, end + 1) if grid[height][c] in "+*"}
    op = ops.pop()

    numbers = []
    for c in range(start, end + 1):
        digits = "".join(
            grid[r][c] for r in range(height)
            if grid[r][c].isdigit()
        )
        if digits:
            numbers.append(int(digits))

    if op == "+":
        answer += sum(numbers)
    elif op == "*":
        prod = 1
        for n in numbers:
            prod *= n
        answer += prod

print(answer)
```

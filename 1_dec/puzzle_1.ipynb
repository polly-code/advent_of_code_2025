```python
class CircleGraph:
    def __init__(self, size=100, start_position=50):
        self.size = size
        self.position = start_position % size

    def move(self, instruction):
        direction = instruction[0]
        steps = int(instruction[1:])

        if direction == 'L':
            self.position = (self.position - steps) % self.size
        elif direction == 'R':
            self.position = (self.position + steps) % self.size
        else:
            raise ValueError("Instruction must start with 'L' or 'R'.")

        return self.position
```

```python
def count_zero_visits(filename):
    graph = CircleGraph()
    zero_count = 0

    with open(filename, 'r') as file:
        instructions = file.readlines()
        for instruction in instructions:
            position = graph.move(instruction.strip())
            if position == 0:
                zero_count += 1

    return zero_count
```

```python
count_zero_visits('input.txt')
```

```python
def count_zero_passes(filename) -> int:
    graph = CircleGraph()
    zero_passes = 0
    with open(filename, 'r') as file:
        instructions = file.readlines()
        for instruction in instructions:
            steps = int(instruction[1:].strip())
            direction = instruction[0]
            s_pos = graph.position
            assert steps != 0
            zero_passes += steps // graph.size
            # check if we passed zero in the remaining steps
            if direction == 'L' and s_pos != 0:
                if (s_pos - steps % graph.size) <= 0:
                    zero_passes += 1
            elif direction == 'R':
                if (s_pos + steps % graph.size) >= graph.size:
                    zero_passes += 1
            graph.move(instruction.strip())
    return zero_passes
```

```python
count_zero_passes('input.txt')
```

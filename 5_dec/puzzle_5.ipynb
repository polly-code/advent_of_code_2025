```python
with open("input.txt") as f:
    input = f.readlines()
```

```python
ranges = []
ids = []
for line in input:
    if line.strip() == '':
        continue
    if '-' in line:
        parts = line.strip().split('-')
        ranges.append((int(parts[0]), int(parts[1])))
    else:
        ids.append(int(line.strip()))
```

```python
total_fresh = 0
for i in ids:
    for r in ranges:
        if r[0] <= i <= r[1]:
            total_fresh += 1
            break
total_fresh
# it's not an efficient way of doing it, but good enough since the input data is so small ^_^
```

```python
sorted_ranges = sorted(ranges, key=lambda x: x[0])
new_ranges = [sorted_ranges[0]]
for i, r in enumerate(sorted_ranges[1:]):
    if r[0] <= new_ranges[-1][1]:
        new_ranges[-1] = (new_ranges[-1][0], max(new_ranges[-1][1], r[1]))
    else:
        new_ranges.append(r)

total_fresh_ids = 0
for r in new_ranges:
    total_fresh_ids += r[1] - r[0] + 1
total_fresh_ids
```
